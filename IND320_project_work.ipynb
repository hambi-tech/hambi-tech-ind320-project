Loading and Display the Data

```python
import pandas as pd

# Load the CSV file
data = pd.read_csv('open-meteo-subset.csv')

# Display the first few rows of the data
data.head()
```

### Clean the column names

```python
# Clean the column names by removing unwanted characters and stripping spaces
data.columns = data.columns.str.replace('Â', '')# Remove unwanted character
data.columns = data.columns.str.strip() # Remove leading/trailing spaces
data.head()
```

###  Over Time

This plot shows the variation in temperature,Precipitation,Wind Speed  over time. 


```python
import matplotlib.pyplot as plt

# Plot Temperature
plt.figure(figsize=(10, 6))
plt.plot(data['time'], data['temperature_2m (°C)'])  
plt.title('Temperature Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.show()

# Plot Precipitation
plt.figure(figsize=(10, 6))
plt.plot(data['time'], data['precipitation (mm)'])
plt.title('Precipitation Over Time')
plt.xlabel('Time')
plt.ylabel('Precipitation (mm)')
plt.xticks(rotation=45)
plt.show()

# Plot Wind Speed
plt.figure(figsize=(10, 6))
plt.plot(data['time'], data['wind_speed_10m (m/s)'])
plt.title('Wind Speed Over Time')
plt.xlabel('Time')
plt.ylabel('Wind Speed (m/s)')
plt.xticks(rotation=45)
plt.show()
```

# Ploting All Columns Together


This plot displays temperature, precipitation, and wind speed on the same graph, allowing for a comparison of all three variables over time.


```python
# Plot all columns together
plt.figure(figsize=(10, 6))
plt.plot(data['time'], data['temperature_2m (°C)'], label='Temperature')
plt.plot(data['time'], data['precipitation (mm)'], label='Precipitation')
plt.plot(data['time'], data['wind_speed_10m (m/s)'], label='Wind Speed')
plt.title('Weather Data Over Time')
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
